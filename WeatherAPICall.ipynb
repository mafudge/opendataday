```python
import requests

key = "" # sign up for your own key at https://openweathermap.org/ 
query_string = { 'lat' : 43.0481221, 'lon': -76.1474244, 'appid' : key, 'units' : 'imperial'}
url = 'https://api.openweathermap.org/data/2.5/onecall'
response = requests.get(url, params=query_string)
response.raise_for_status()
weather = response.json()
weather
```
